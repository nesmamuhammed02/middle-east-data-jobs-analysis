```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
import ast
from datasets import load_dataset

# Loading Data
# The dataset is pulled directly from Hugging Face for reproducibility
ds = load_dataset("lukebarousse/data_jobs")
# Convert the train split to a pandas DataFrame for easier manipulation
df=ds['train'].to_pandas()


# Data Cleanup
# Convert job_posted_date from string to datetime for time-based operations
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# job_skills is stored as a string representation of a list — parse it back to an actual list
# Non-null values are evaluated using ast.literal_eval; null values are left as-is
df['job_skills']= df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

```

## Market Overviw

```python
# Define the list of Middle East countries to filter the dataset
middle_east=['Algeria', 'Bahrain', 'Egypt', 'Iraq', 'Jordan', 'Kuwait', 'Lebanon', 'Libya', 'Morocco', 'Oman', 'Palestine', 'Qatar', 'Saudi Arabia', 'Tunisia', 'Turkey', 'United Arab Emirates','Yemen']

# Filter the dataset to include only rows where the job country is in the Middle East
df_me=df[df['job_country'].isin(middle_east)]
df_me.info()
```

```python
# Count job postings per country and keep the top 10
df_top_countries = df_me['job_country'].value_counts().head(10).to_frame() 

# Plot a horizontal bar chart of the top 10 countries
sns.set_theme(style="whitegrid")
sns.barplot(data=df_top_countries ,x='count',y='job_country',hue='job_country', palette="dark:purple",legend=False)
sns.despine()
plt.title('Top 10 Countries with the Most Data Jobs in the Middle East', fontsize=16,loc='right')
plt.xlabel('Number of Data Jobs')
plt.ylabel('')
plt.show()
```

##### The UAE clearly leads the region, standing out from the other countries.Saudi Arabia and Egypt follow, showing strong but noticeably lower demand for data jobs.


```python
# Count job postings per company and keep the top 10
df_top_companies = df_me['company_name'].value_counts().head(10).to_frame() 

# Plot a horizontal bar chart of the top 10 hiring companies
sns.set_theme(style="whitegrid")
sns.barplot(
    data=df_top_companies,
    x='count',
    y='company_name',
    hue='company_name',
    palette="dark:purple",
    legend=False
    )
sns.despine()
plt.title('Top 10 Companies with the Most Data Jobs in the Middle East', fontsize=16,loc='right')
plt.xlabel('Number of Data Jobs')
plt.ylabel('')
plt.show()
```

##### Talent Pal clearly leads the region, standing out as the top employer for data professionals.    
##### Following Talent Pal, other companies also offer significant opportunities.


```python
# Count occurrences of each job title and reset the index for plotting
df_top_data_jobs= df_me['job_title_short'].value_counts()
df_top_data_jobs=df_top_data_jobs.reset_index(name='job count')

# Highlight the top 3 roles with a darker color to draw attention
n=len(df_top_data_jobs)
colors = ['#520c61'] * 3 + ['#982286'] * (n - 3) 


sns.barplot(
    data = df_top_data_jobs,
    x='job_title_short',
    y='job count',
    palette=colors,
    )

plt.xticks(rotation=45, ha='right')
plt.title('Top In-Demand Data Roles in the Middle East', fontsize=16,loc='center')
plt.xlabel('')
plt.ylabel('Number of Job Postings')

sns.despine()
plt.show()

```

##### The previous analysis revealed that Data Analyst, Data Scientist, and Data Engineer are the most in-demand roles in the Middle East job market.
